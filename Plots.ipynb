```python
# Data
data = [
    ['Mahesh', 35],
    ['Utkarsha', 30],
    ['Gayathri', 65],
    ['Antariksha', 85],
    ['Abhra', 25],
    ['Yash', 50],
    ['X', 40],
    ['Y', 30],
    ['Z', 66]
]

# Extract values
names = [item[0] for item in data]
values = [item[1] for item in data]

# Plotting the box plot
plt.boxplot(values)

# Labelling the x-axis with names
plt.xticks(range(1, len(names) + 1), names)

# Labelling the diagram
plt.xlabel('Names')
plt.ylabel('Values')

# Display the box plot
plt.show()
```
